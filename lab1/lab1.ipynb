```python
#2,4
def operacje_macierze(size, A, B):
    if A.shape != (size, size) or B.shape != (size, size):
        raise ValueError("Złe rozmiary macierzy")
    print("Macierz odwrotna A:\n", np.linalg.inv(A))
    print("Transpozycja macierzy B:\n", B.T)
    print("Suma A i B:\n", A+B)
    print("Różnica A i B:\n", A-B)
    print("Iloczyn A i B:\n", np.dot(A,B))
    print("Mnożenie A przez 10:\n", 10*A)

n = 3
operacje_macierze(n, np.random.rand(n,n),np.random.rand(n,n))
    
```

```python
#2.5

import time 

def cos_time():
    x = np.arange(-10, 10, 0.001)
    
    # petla
    start_time = time.time()
    y = [np.cos(x) for v in x]
    time_result = time.time() - start_time
    print(f"Czas w pętli: {time_result:.6f}s")
    
    # tablicowo
    start_time = time.time()
    y = np.cos(x)
    time_result = time.time() - start_time
    print(f"Czas tablicowy: {time_result:.6f}s")
    
cos_time()
```

```python
#4.4
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(1, 11)
y = np.arange(1, 11)
X, Y = np.meshgrid(x, y)
Z = (X - 5) ** 2 - (Y - 5) ** 2

plt.figure()
plt.contour(X, Y, Z)
plt.title('Contur')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, 50)
ax.set_title('Contur3D')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Surf')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_title('Mesh')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(y)):
    ax.plot(x, Z[i, :], y[i])
ax.set_title('Waterfall')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
#5.5
from scipy.interpolate import interp1d

x = np.arange(-10, 10.5, 0.5)
y = np.cos(3*x) + np.cos(x)

x_dense = np.linspace(-10, 10, 1000)

linear_interp = interp1d(x, y, kind='linear')
cubic_interp = interp1d(x, y, kind='cubic')
nearest_interp = interp1d(x, y, kind='nearest')

polyfits = {}
for deg in [3, 5, 7]:
    coeffs = np.polyfit(x, y, deg)
    polyfits[deg] = np.polyval(coeffs, x_dense)

plt.figure(figsize=(12, 7))
plt.plot(x, y, 'ko', label='Węzły interpolacji', markersize=4)
plt.plot(x_dense, linear_interp(x_dense), label='liniowa')
plt.plot(x_dense, cubic_interp(x_dense), label='sześcienna')
plt.plot(x_dense, nearest_interp(x_dense), label='ns')

for deg, y_fit in polyfits.items():
    plt.plot(x_dense, y_fit, '--', label=f'polyfit (stopień {deg})')

plt.title("Interpolacja")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
